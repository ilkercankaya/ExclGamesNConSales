```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('Video_Game_Sales_as_of_Jan_2017.csv')
dataset
```

```python
dataset.dropna(inplace=True)
dataset = dataset.loc[dataset['Publisher'].isin(['Nintendo','Sony Computer Entertainment','Microsoft Game Studios'])]
dataset = dataset[dataset['Year_of_Release'] > 2007.0]
dataset = dataset[dataset['Global_Sales'] > 0.1]
dataset.drop(['Name','Rating','Year_of_Release', 'Other_Sales', 'JP_Sales', 'EU_Sales', 'NA_Sales','Critic_Count','User_Count'], axis=1, inplace=True)
dataset
```

```python
def hit_check(x):
    if x >= 0.7:
        return 1
    else:
        return 0
#Set the 1 or 0 according to Global sales more than 700.000
dataset['Hit'] = dataset['Global_Sales'].apply(hit_check)
dataset.drop('Global_Sales',inplace=True,axis=1)
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import metrics

le = LabelEncoder()
#take object type columns to transform
cont_attributes = dataset.select_dtypes(exclude=['object'])
obj_attributes = dataset.select_dtypes(include=['object'])
```

```python
#Transform the type of columns object to numeric
obj_attributes_labelled = obj_attributes.apply(le.fit_transform)
merged_label = pd.concat([cont_attributes,obj_attributes_labelled],axis=1)
merged_label
```

```python
#Set the feature and target
features = merged_label.drop(['Hit'], axis=1)
target = merged_label[['Hit']]
target = np.ravel(target)
```

```python
#Split the data
X_train, X_test, Y_train, Y_test = train_test_split(features, target, test_size=0.35)
```

## Vector Machines Model

```python
from sklearn.svm import SVC
```

```python
#use SVC model
svc_model = SVC()
svc_model.fit(X_train,Y_train)
```

```python
svc_predict = svc_model.predict(X_test)
expected = Y_test

# Compare results
print ("Accuracy: " , metrics.accuracy_score(expected,svc_predict))
```

```python
# Import all SVM 
from sklearn import svm

# SVM regularization parameter
C = 1

# SVC with a Linear Kernel  (our original example)
svc = svm.SVC(kernel='linear', C=C).fit(X_train, Y_train)

# Gaussian Radial Bassis Function
rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X_train, Y_train)

# SVC with 3rd degree poynomial
poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(X_train, Y_train)

# SVC Linear
lin_svc = svm.LinearSVC(C=C).fit(X_train, Y_train)
```

```python
prediction_svc =svc.predict(X_test)
print("linear kernel accuracy: " , metrics.accuracy_score(Y_test,prediction_svc))
prediction_rbf =svc.predict(X_test)
print("Gausian radial bassis accuracy: " , metrics.accuracy_score(Y_test,prediction_rbf))
prediction_poly =svc.predict(X_test)
print("Polynomial SVC accuracy: " , metrics.accuracy_score(Y_test,prediction_poly))
prediction_lin =svc.predict(X_test)
print("linear SVC accuracy: " , metrics.accuracy_score(Y_test,prediction_lin))
```
